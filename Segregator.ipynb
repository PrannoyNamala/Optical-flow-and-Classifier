```python
# Run this to make a Resized GT dataset

import os
import shutil
import cv2
current_path = '/content/gdrive/MyDrive/Classification Project/Fish_Dataset/Fish_Dataset'
moving_path = '/content/gdrive/MyDrive/Classification Project/Fish_Dataset/Resized_Dataset_GT'
if os.path.exists(moving_path):
    shutil.rmtree(moving_path)
    os.makedirs(moving_path)
for folder in os.listdir(current_path):
    print("\n")
    print("Generating Training & Validation set for " + folder, end='')
    moving_training_folder = moving_path + '/train/' + folder
    moving_validation_folder = moving_path + '/validation/' + folder
    os.makedirs(moving_training_folder)
    os.makedirs(moving_validation_folder)
    number_of_files = len([f for f in os.listdir(current_path + '/' + folder+'/'+folder  + ' GT')])
    counter = -1
    for file in os.listdir(current_path + '/' + folder+'/'+folder  + ' GT'):
        print(".", end='')
        counter = counter + 1
        if counter <= 0.8 * number_of_files:
            image = cv2.imread(current_path + '/' + folder + '/'+folder  + ' GT'+ '/' + file)
            image = cv2.resize(image, (224, 224))
            file_name = moving_training_folder+'/'+file
            cv2.imwrite(file_name, image)
        if counter > 0.8 * number_of_files:
            image = cv2.imread(current_path + '/' + folder+'/'+folder  + ' GT' + '/' + file)
            image = cv2.resize(image, (224, 224))
            file_name = moving_validation_folder + '/' + file
            cv2.imwrite(file_name, image)
print("\n\nDirectories Created!")
```
